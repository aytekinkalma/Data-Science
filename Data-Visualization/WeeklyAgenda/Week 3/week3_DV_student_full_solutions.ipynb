```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

```python
sns.set_style("darkgrid")
```

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

```python
df.plot.bar();
```

```python
df.plot(kind = "bar", figsize =(6,4));
```

# 3. Stack the bars

```python
ax = df.plot.bar(stacked = False)
sns.set_style(style = "darkgrid")
plt.xticks(rotation=90)
plt.show()
```

```python
df.plot.bar(stacked=True)
plt.grid(True);
```

```python
ax = df.plot.bar(stacked = True)
sns.set_style(style = "darkgrid")
plt.xticks(rotation=90)
plt.show()
```

# 3. Rotate the labels and set figsize

```python
df.plot.bar(stacked=True, rot = 0, figsize=(9,6));
```

```python
ax = df.plot.bar(figsize = (9,6), rot=0, stacked = True)
sns.set_style(style = "darkgrid")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,6))

ax.bar(df.index, df.income, label = 'income')
ax.bar(df.index, df.expense, label = 'expense')
plt.legend()
plt.xticks(rotation=0);
```

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]

g = df.plot.bar(figsize=(9,6), rot=0)
g.set_xticklabels(labels)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.03, p.get_height()+0.5))
```

# 7. Plot the avg tip by size  (lineplot)

```python
tips.groupby('size').tip.mean().plot();
```

```python
sns.lineplot(data=tips, x='size', y='tip');
```

```python
sns.lineplot(data=tips, x='size', y='tip', ci=None);
```

```python
size_by_average = tips.groupby("size")["tip"].mean()
plt.plot(size_by_average.index, size_by_average.values)
plt.xlabel("size")
plt.show()
```
